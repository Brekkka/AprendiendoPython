8. Make a function that takes a categorical series and a number (n) and returns a replace series that replaces any value not in the top n most frequent values with 'Other'

```python
def replace_non_top(Series, n):
    Series_cat = Series.astype('category')
    top_n = Series_cat.value_counts().index[:n]
    if 'Other' in Series_cat.cat.categories:
        return Series_cat.where(Series_cat.isin(top_n), other = 'Other')
    else:
        Series_cat.cat.add_categories('Other')
        return where(Series_cat.isin(top_n), other='Other')
```

```python
replace_non_top(countries, 2)
```
